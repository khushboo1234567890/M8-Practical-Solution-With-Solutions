```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/download (1).png")

if image is None:
    print("Error: 'image1.jpg' not found. Please ensure the image file is uploaded to your Colab environment or provide the correct path.")
else:
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")
    plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Threshold
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(threshold, cmap='gray')
plt.title("Threshold Image")
plt.axis("off")
plt.show()
```

```python
blur = cv2.GaussianBlur(image, (5,5), 0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

```python
median = cv2.medianBlur(image, 5)

plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title("Median Filter")
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

```python
orb = cv2.ORB_create()

keypoints, descriptors = orb.detectAndCompute(gray, None)

feature_image = cv2.drawKeypoints(
    image,
    keypoints,
    None,
    color=(0,255,0)
)

plt.imshow(cv2.cvtColor(feature_image, cv2.COLOR_BGR2RGB))
plt.title("Feature Detection")
plt.axis("off")
plt.show()
```
